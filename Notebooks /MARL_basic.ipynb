Will be using OpenAi gym to build out the environment

```python
environment_name = 'CartPole-v1'
env = gym.make(environment_name) # gym.make is a function, which initializes the environment, meaning creates an environment.
```

here I created a CartPole-v1 environement and stored it in env variable, create the environment using gym.make() function

```python
episodes = 20
x = []
y=[]
for episode in range(1, episodes+1):
   state = env.reset()
   done = False
   score = 0
   while not done:
      env.render()
      action = env.action_space.sample() # sample() to pick one action_space out of 2. for Cartpole it creates only two action spaces (0 to move left, 1 to move right)
      n_state, reward, done, info, *extra = env.step(action) # env.step() is going to give output includes n_state(next state, based on present state, reward, done, info)
      score += reward
  
   x.append(episode)
   y.append(score)
  
env.close()


```
